```python
import pandas as pd
import numpy as np
import streamlit as st
```

```python
df = pd.read_csv('athlete_events.csv')
region_df = pd.read_csv('noc_regions.csv')
```

```python
df.sample(5)
```

```python
df.shape[0]
```

```python
df.shape
```

```python
df = df[df['Season']=="Summer"]
```

```python
df.sample(5)
```

```python
df = df.merge(region_df, on='NOC', how='left')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["Medal"].value_counts()
```

```python
df1 = pd.get_dummies(df["Medal"])
df1
```

```python
df.shape
```

```python
df = pd.concat([df,df1], axis=1)
```

```python
df
```

```python
df.groupby(df['NOC']).sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```

```python
# Team	NOC	Games	Year	Season	City	Sport	Event	Medal
medal_tally = df.drop_duplicates(subset=['Team','NOC','Games','Year','Season','City','Sport','Event','Medal'])
```

```python
medal_tally
```

```python
medal_tally = medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```
